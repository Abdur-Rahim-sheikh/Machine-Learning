```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('melbourne_housing.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(['Lattitude','Longtitude','YearBuilt','Postcode','Date','Address'],axis='columns',inplace=True)
```

```python
df.isna().sum()
```

```python
df[['Car','Bedroom2','Bathroom']] = df[['Car','Bedroom2','Bathroom']].fillna(0)
```

```python
df.Landsize.fillna(df.Landsize.mean(),inplace=True)
df.BuildingArea.fillna(df.BuildingArea.mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df2 = pd.get_dummies(df,drop_first=True)
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.drop(['Price'],axis=1),df2['Price'],random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model.intercept_
```

```python
from sklearn.linear_model import Lasso
model2 = Lasso(alpha=50,max_iter=500,tol=0.001)
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```

```python
from sklearn.linear_model import Ridge
model3 = Ridge(alpha=50,max_iter=100,tol=0.001)
model3.fit(X_train,y_train)
model3.score(X_test,y_test)
```
