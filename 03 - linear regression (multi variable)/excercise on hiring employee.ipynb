```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.rename(columns = {'test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score','salary($)':'salary'},inplace = True)
df
```

```python
df.experience.fillna('three',inplace=True)
df
```

```python
from word2number import w2n

df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
df.test_score.fillna(df.test_score.median(),inplace=True)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop(['salary'],axis='columns'),df.salary)
```

```python
reg.predict([[2,9,6]])
```
